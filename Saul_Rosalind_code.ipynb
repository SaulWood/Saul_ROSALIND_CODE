```python
def at_least_one_dom_allele(k, m, n):
    prob_no_dom_allele=(n/(k+m+n))*((n-1)/(k+m+n-1)) + 0.5*(n/(k+m+n))*(m/(k+m+n-1)) + 0.25*(m/(k+m+n))*((m-1)/(k+m+n-1)) + 0.5*(m/(k+m+n))*(n/(k+m+n-1))
    prob_dom_allele=1-prob_no_dom_allele
    return prob_dom_allele
```

```python
def gc_content(base_sequence):
   gc_count=0
   for base in base_sequence:
      if base == 'G' or base=='C':
         gc_count = gc_count +1
   percentage_GC = (gc_count/len(base_sequence)) * 100 
   return percentage_GC
```

```python
def number_of_combinations_possible(polypeptide_sequence):
    num_codons_list=[]
    aa_to_num_codons_dict={'M':1,'F':2,'L':6,'I':3,'V':4,'S':6,'P':4,'T':4,'A':4,'Y':2,'H':2,'Q':2,'N':2,'K':2,'D':2,'E':2,'C':2,'W':1,'R':6,'G':4 }
    # There will always be a stop codon at end of polypetide. 
    # But to ensure stop codon variability is accounted for in number of potential RNA strings I just multiply by 3. 
    # Have found number of combinations of RNA strings in the pp chain itself by looping through the polypeptide and using the dictionary to find the number of possible 
    # codons for each aa and then multiplying them all together to conert this info into number of potential RNA strings that could have generated it.
    # That final value is mod 1*10^6 to give computer a more manageable number to store (reduces space needed).

    for aa in polypeptide_sequence:
        num_codons_list.append(aa_to_num_codons_dict[aa])
    
    return num_codons_list
```
